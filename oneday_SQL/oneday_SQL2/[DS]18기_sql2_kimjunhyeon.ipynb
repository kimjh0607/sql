```python
# 주소를 입력받아 구 이름을 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)
def get_gu(address):
    gu = address.split()[1]

    return gu
```

```python
# 주소를 입력받아 위도, 경도를 반환하는 함수 (테스트 입력 : ‘서울시 강남구 헌릉로 730’)
import googlemaps
gmaps_key = "AIzaSyAgvvviPD4RiDnMZ6FTBI67zGprcnSyV-0"

def get_lat_lng(address):
    gmaps = googlemaps.Client(gmaps_key)

    lat = gmaps.geocode(address)[0].get("geometry")["location"]["lat"]
    lng = gmaps.geocode(address)[0].get("geometry")["location"]["lng"]
    
    return lat, lng
```
