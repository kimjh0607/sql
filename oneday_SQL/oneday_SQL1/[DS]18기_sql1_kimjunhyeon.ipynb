시각화 프로젝트를 위하여 다음의 규칙으로 쿼리하여 CSV 파일로 저장합니다. (Python 코드로 작업)

- 전체 데이터를 가져오는데, 각 스타벅스 매장별로 이디야 전체 매장정보가 매칭되어 있어야 합니다. (정렬 : s_id, e_id 순)
- 다음의 형식으로 저장되어야 합니다. (브랜드 이름, 칼럼 명 주의)
- 데이터 프레임 출력을 해주세요. 데이터 프레임 미출력시 감점입니다.

제출 8.
- 시각화 프로젝트 관련 코드 (ipynb), 결과 파일 (csv)

```python
import pandas as pd

sql = "select s.id s_id, b.name s_brand, s.name s_name, s.gu_name s_gu, s.address s_address, s.lat s_lat, s.lng s_lng  from COFFEE_STORE s, COFFEE_BRAND b where b.id= s.brand and s.brand=1 order by s_id"
cur.execute(sql)
result_sb = cur.fetchall()

sql = "select s.id e_id, b.name e_brand, s.name e_name, s.gu_name e_gu, s.address e_address, s.lat e_lat, s.lng e_lng  from COFFEE_STORE s, COFFEE_BRAND b where b.id = s.brand and s.brand=2 order by e_id "
cur.execute(sql)
result_ed = cur.fetchall()

# 스타벅스
sb_columns = ["s_id", "s_brand", "s_name", "s_gu", "s_address", "s_lat", "s_lng" ]
starbucks_df = pd.DataFrame(result_sb, columns=sb_columns)

# 이디야
ed_columns = ["e_id", "e_brand", "e_name", "e_gu", "e_address", "e_lat", "e_lng" ]
ediya_df = pd.DataFrame(result_ed, columns=ed_columns)

# 합치기
sb_ed_all = pd.concat([starbucks_df, ediya_df], axis=1)
```

```python
# 0으로 채운 후 int로 바꾸기
sb_ed_all["s_id"] = sb_ed_all["s_id"].fillna(0)
sb_ed_all["s_id"] = sb_ed_all["s_id"].astype(int)
```

```python
# 인덱스 변경
sb_ed_all.set_index("s_id", inplace=True)
```

```python
# csv파일로 저장
sb_ed_all.to_csv("./total_cafe_info.csv", sep=",", encoding="utf-8")

conn.close()

cafe_info_rows = pd.read_csv("./total_cafe_info.csv", index_col=0, encoding="utf-8")
cafe_info_rows.head()
```
