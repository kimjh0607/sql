#### cursor 만들기
- 읽어올 양이 많은 경우 buffer 설정을 해준다.

```python
cursor = conn.cursor(buffered=True)
```

insert문 만들기

```python
sql = 'insert into police_station values (%s, %s)'
```

#### 데이터 입력
- commit()은 database에 적용하기 위한 명령어

```python
for i, row in df.iterrows():
    cursor.execute(sql, tuple(row))
    print(tuple(row))
    conn.commit()
```

```python
cursor.execute('select * from police_station')

result = cursor.fetchall()
result[:5]
```

#### 검색결과를 pandas로 읽기

```python
df = pd.DataFrame(result)
df.tail()
```

# Python with CSV 예제
- crime_status 테이블에 2020_crime.csv 를 입력하는 코드를 작성

```python
import mysql.connector

conn = mysql.connector.connect(
    host = 'database-1.chmbp2zmsp5d.ap-southeast-2.rds.amazonaws.com',
    port = 3306,
    user = 'zero',
    password = 'zerobase',
    database = 'zerobase'
)
```

#### 2020_crime.csv 데이터 읽어오기 (encoding="euc-kr")

```python
import pandas as pd

df = pd.read_csv('2020_crime.csv', encoding='euc-kr')
df.head()
```

#### insert Query작성

```python
sql = 'insert into crime_status values("2020", %s, %s, %s, %s)'
cursor = conn.cursor(buffered=True)
```

```python
for i, row in df.iterrows():
    cursor.execute(sql, tuple(row))
    print(tuple(row))
    conn.commit()
```

```python
df.tail()
```

#### crime_status 테이블의 데이터 조회

```python
cursor.execute("select * from crime_status")

result = cursor.fetchall()

for row in result:
    print(row)
```

#### 조회한 결과를 pandas로 변환해서 확인

```python
df = pd.DataFrame(result)
df.head()
```
